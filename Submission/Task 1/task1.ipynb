```python
def get_bus_indexes(df)->list:

    mean_bus = df['bus'].mean()

    bus_indexes = df[df['bus'] > 2 * mean_bus].index.tolist()
    bus_indexes.sort()
    return bus_indexes
    
```

```python
def filter_routes(df)->list:


    mean_truck = df.groupby('route')['truck'].mean()


    filtered_routes = mean_truck[mean_truck > 7].index.tolist()


    filtered_routes.sort()

    return filtered_routes


filter_routes(dataset)
```

```python
def multiply_matrix(matrix)->pd.DataFrame:
    conditions = [
        (matrix > 20),
        ( matrix<= 20)
    ]

    factors = [0.75, 1.25]

    matrix = np.select(conditions, [matrix * factor for factor in factors], matrix)
    matrix = np.round(matrix, 1)

    return pd.DataFrame(matrix)

```
